# Les Calcules Statistiques

```python
np.corrcoef(angle)
```

```python
np.unique(angle, return_counts=True)
```

```python
val = np.unique(angle, return_counts=True)[0]
occurs = np.unique(angle, return_counts=True)[1]
print(val)
print(occurs)
print(val[np.argsort(occurs)])
print(occurs[np.argsort(occurs)])
```

```python
angle
```

```python
for valeur, occurence in zip(val[np.argsort(occurs)][::-1], occurs[np.argsort(occurs)][::-1]):
    print('la valeur: {} apparaît {} fois'.format(valeur, occurence))
```

# Exercice

Zoom d'un l'image face de misc

```python
from scipy import misc
import matplotlib.pyplot as plt
```

```python
face = misc.face()
plt.imshow(face)
plt.show()
```

```python
face = misc.face(gray=True)
face = face[::4, ::4]
plt.imshow(face, cmap=plt.cm.gray)
plt.show()
```

```python
heigth = face.shape[0]
width = face.shape[1]
```

```python
heigth
```

```python
width
```

```python
zoom = face[heigth//4:-heigth//4, width//4:-width//4]
```

```python
plt.imshow(zoom, cmap=plt.cm.gray)
plt.show()
```

```python
zoom[zoom>=150] = 255
plt.imshow(zoom)
plt.show()
```
