```python
f = lambda x, y: np.sin(x) + np.cos(x+y)
```

```python
x = np.linspace(0,5, 10)
y = np.linspace(0, 5, 100)

X, Y = np.meshgrid(x, y)
```

```python
Z = f(X, Y)
```

```python
ax = plt.axes(projection="3d")
ax.plot_surface(X, Y, Z, cmap="plasma")
```

```python
dataset = {f'experiece {i}': np.random.randn(100, 3) for i in range(7)}
```

```python
def graphique(data):
    n = len(data)
    plt.figure(figsize=(12, 16))
    for k, i in zip(data.keys(), range(1, n+1)):
        plt.subplot(n, 1, i)
        plt.plot(data[k], label="la sinuso√Ødale")
        plt.title(k)
        plt.xlabel('Axe X')
        plt.ylabel('Axe Y')
        plt.legend()
    plt.show()
```

```python
graphique(dataset)
```
