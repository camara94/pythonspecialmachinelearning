```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.array([1, 2, 3, 4])
Y = np.array([y**2 for y in X])
```

```python
X
```

```python
Y
```

```python
plt.plot(X, Y)
plt.show()
```

```python
plt.scatter(X, Y)
plt.show()
```

```python
plt.bar(X, 220)
```

```python
plt.figure()
plt.plot(X, Y, c='green', lw=5, ls='--', label="quandratique")
plt.plot(Y, X, c='brown', lw=4, ls='--', label='lineaire')
plt.xlabel('Axe X')
plt.ylabel('Axe Y')
plt.title('Figure 1')
plt.legend()
```

```python
plt.figure()
plt.subplot(2,2, 1)
plt.plot(np.array([x for x in range(5)]), np.array([x**3 for x in range(5)]))
plt.subplot(2,2, 2)
plt.plot(np.array([np.sin(x) for x in range(5)]), np.array([np.cos(x**3) for x in range(5)]), c='red')
plt.subplot(2,2, 3)
plt.plot(np.array([x for x in range(5)]), np.array([x**3 for x in range(5)]))
plt.subplot(2,2, 4)
plt.plot(np.array([x for x in range(5)]), np.array([x**3 for x in range(5)]))
```
